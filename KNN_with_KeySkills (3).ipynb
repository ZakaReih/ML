```python
import pandas as pd

# 1. Обработка пропусков
df['description'] = df['description'].fillna('Не указано')
df['requirements'] = df['requirements'].fillna('Не указано')
df['key_skills'] = df['key_skills'].apply(lambda x: ' '.join(x) if isinstance(x, list) else 'Не указано')
df['companyName'] = df['companyName'].fillna('Не указано')
df['salary_currency'] = df['salary_currency'].fillna('Не указано')

# 2. Заполнение зарплат с учетом медианы
df['salary_from'] = pd.to_numeric(df['salary_from'], errors='coerce')
df['salary_to'] = pd.to_numeric(df['salary_to'], errors='coerce')

df['salary_from'].fillna(df['salary_to'], inplace=True)
df['salary_to'].fillna(df['salary_from'], inplace=True)

salary_median = df[['salary_from', 'salary_to']].median().median()
df['salary_from'].fillna(salary_median, inplace=True)
df['salary_to'].fillna(salary_median, inplace=True)

# 3. Очистка текстовых данных
def clean_text(text):
    text = text.lower()
    text = ''.join([char for char in text if char.isalpha() or char.isspace()])
    return text

df['description_clean'] = df['description'].apply(clean_text)
df['requirements_clean'] = df['requirements'].apply(clean_text)
df['key_skills_clean'] = df['key_skills'].apply(clean_text)

# 4. Проверка дубликатов перед удалением
before = df.shape[0]
print("Дубликатов перед удалением:", df.duplicated(subset=['title', 'companyName', 'salary_from', 'requirements']).sum())

# 5. Удаление дубликатов (учитываем зарплату и требования)
if 'published_at' in df.columns:
    df.sort_values(by='published_at', ascending=False, inplace=True)  # Оставляем самую свежую вакансию

df.drop_duplicates(subset=['title', 'companyName', 'salary_from', 'requirements'], keep='first', inplace=True)

after = df.shape[0]
print(f"Удалено {before - after} строк (из {before})")

# 6. Анализ данных
salary_stats = df[['salary_from', 'salary_to']].describe()
print("Статистика по зарплатам:")
print(salary_stats)
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
from nltk.corpus import stopwords

# Стоп-слова для удаления ненужных слов
stop_words = set(stopwords.words("russian"))

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zа-я0-9#+-]', ' ', text)
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)

# Заполнение NaN и обработка текста
df['requirements'] = df['requirements'].fillna('')
df['job_text'] = df['requirements'] + ' ' + df['conditions']

# Функция для поиска вакансий и недостающих навыков
def recommend_jobs_ml(student_skills, job_type=None, location=None, top_n=5):
    student_skills = set(student_skills.lower().split())  # Преобразуем в множество

    # Фильтрация по условиям занятости и локации
    filtered_df = df.copy()
    if job_type:
        filtered_df = filtered_df[filtered_df['conditions'].str.contains(job_type, case=False, na=False)]
    if location:
        filtered_df = filtered_df[filtered_df['locations'].str.contains(location, case=False, na=False)]

    if filtered_df.empty:
        return "Ничего не найдено.", set()

    # Векторизация
    vectorizer = TfidfVectorizer()
    X_filtered = vectorizer.fit_transform(filtered_df['job_text'])  # Матрица вакансий
    student_vector = vectorizer.transform([" ".join(student_skills)])  # Вектор студента

    # Обучение KNN
    knn = NearestNeighbors(n_neighbors=min(top_n, len(filtered_df)), metric='cosine')
    knn.fit(X_filtered)

    # Поиск ближайших вакансий
    distances, indices = knn.kneighbors(student_vector)

    # Добавляем схожесть (1 - дистанция, чтобы было понятнее)
    recommended_jobs = filtered_df.iloc[indices[0]].copy()
    recommended_jobs["similarity_score"] = 1 - distances[0]  # Чем ближе к 1, тем лучше

    # Сортируем по схожести
    recommended_jobs = recommended_jobs.sort_values(by="similarity_score", ascending=False)

    # Проверяем, есть ли хоть одна вакансия, где встречаются навыки студента
    valid_jobs = []
    for _, row in recommended_jobs.iterrows():
        job_skills = set(row['requirements'].lower().split())
        if job_skills & student_skills:  # Если есть хотя бы одно совпадение
            valid_jobs.append(row)

    if not valid_jobs:
        return "Ничего не найдено.", set()

    valid_jobs_df = pd.DataFrame(valid_jobs)  # Преобразуем в DataFrame

    # Вычисляем недостающие навыки (не сокращенные)
    all_required_skills = set()
    for req in valid_jobs_df["requirements"]:
        all_required_skills.update(req.lower().split())  # Разбиваем на слова

    missing_skills = all_required_skills - student_skills

    return valid_jobs_df[['title', 'locations', 'similarity_score']], missing_skills

# Пример
student_skills = "sql ml"
job_type = "Полная занятость"
location = "Алматы"
result, missing_skills = recommend_jobs_ml(student_skills, job_type, location)

print(result)  # Вывод вакансий
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

# Загрузка данных
df = pd.read_csv("updated_file.csv")  # 'key_skills_y' должен быть с навыками через запятую
df["key_skills_y"] = df["key_skills_y"].fillna("")

# Обучение TF-IDF
vectorizer = TfidfVectorizer()
df["key_skills_for_tfidf"] = df["key_skills_y"].apply(lambda x: " ".join([s.strip().lower() for s in x.split(",")]))
vectorizer.fit(df["key_skills_for_tfidf"])

# Функция нормализации навыков
def normalize_skill(skill):
    return skill.strip().lower()

# Удаление дубликатов и близких навыков
def remove_similar_skills(missing_skills, student_skills_list):
    cleaned_skills = set(missing_skills)
    for skill in missing_skills:
        for s_skill in student_skills_list:
            if s_skill in skill or skill in s_skill:
                cleaned_skills.discard(skill)
                break
    return cleaned_skills

# Главная функция рекомендации
def recommend_jobs_ml(student_skills, job_type=None, location=None, top_n=5):
    student_list = [normalize_skill(s) for s in student_skills.split()]
    student_set = set(student_list)

    # Подготовка TF-IDF
    df["key_skills_for_tfidf"] = df["key_skills_y"].apply(lambda x: " ".join([normalize_skill(s) for s in x.split(",")]))
    X = vectorizer.transform(df["key_skills_for_tfidf"])
    student_vector = vectorizer.transform([" ".join(student_list)])

    # Поиск похожих
    knn = NearestNeighbors(n_neighbors=top_n, metric="cosine")
    knn.fit(X)
    distances, indices = knn.kneighbors(student_vector)

    results = []
    all_missing = set()
    seen_urls = set()

    for i, idx in enumerate(indices[0]):
        row = df.iloc[idx]
        url = row["vacancy_url"]

        if url in seen_urls:
            continue  # пропуск дубликатов
        seen_urls.add(url)

        job_skills = set([normalize_skill(s) for s in row["key_skills_y"].split(",") if s.strip()])
        missing_skills = job_skills - student_set
        filtered_missing = remove_similar_skills(missing_skills, student_list)

        all_missing.update(filtered_missing)

        results.append({
            "title": row["title"],
            "locations": row["locations"],
            "similarity_score": round(1 - distances[0][i], 2),
            "vacancy_url": url,
            "missing_skills": ", ".join(filtered_missing)
        })

    results_df = pd.DataFrame(results).sort_values(by="similarity_score", ascending=False)

    return results_df, all_missing

# Пример использования
student_skills = "css html git react javascript"
job_type = "Полная занятость"
location = "Алматы"

result_df, missing_skills = recommend_jobs_ml(student_skills, job_type, location, top_n=10)

print(result_df)
print("Недостающие навыки:", missing_skills)
```

```python
import pickle
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

# Проверяем, есть ли нужный столбец
if "key_skills_y" not in df.columns:
    raise ValueError("Столбец 'key_skills_y' отсутствует в файле.")

# Заполняем NaN и преобразуем текст
df["key_skills_y"] = df["key_skills_y"].fillna("").str.lower()

# Обучаем TF-IDF векторизатор
vectorizer = TfidfVectorizer()
X_filtered = vectorizer.fit_transform(df["key_skills_y"])

# Обучаем KNN
knn = NearestNeighbors(n_neighbors=5, metric="cosine")
knn.fit(X_filtered)

# Сохраняем TF-IDF векторизатор
with open("tfidf_vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)

# Сохраняем KNN модель
with open("knn_model.pkl", "wb") as f:
    pickle.dump(knn, f)

print("Модель сохранена успешно!")
```
